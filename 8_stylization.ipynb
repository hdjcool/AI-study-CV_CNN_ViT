```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
style_transfer = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
import matplotlib.pyplot as plt
```

```python
s = plt.imread('y.jpg')
plt.figure(figsize=(10,10))
plt.imshow(s)
```

```python
iu = plt.imread('guga.jpg')
plt.figure(figsize=(10,10))
plt.imshow(iu)
```

```python
style = plt.imread('y.jpg').astype('float32')
style = tf.constant((style / 255)[tf.newaxis])
style = tf.image.resize(style, (256, 256))
```

```python
content = plt.imread('guga.jpg').astype('float32')
content = tf.constant((content / 255)[tf.newaxis])
```

```python
im = style_transfer(content, style)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(im[0][0])
```

```python
plt.imsave('guga_art.jpg', im)
```
